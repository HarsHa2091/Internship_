```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import requests
import time
from selenium.common.exceptions import NoSuchElementException , StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
driver1=webdriver.Chrome()
```

```python
url1="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver1.get(url1)
time.sleep(2)
```

```python
Name=[]
Artist=[]
Upload_date=[]
Views=[]

nm = driver1.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]/a')
for name in nm:
    Name.append(name.text)

try:
    art = driver1.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]/a')
    for artist in art:
        Artist.append(artist.text)
except NoSuchElementException:
    Artist.append('NA')

up = driver1.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for upload in up:
    Upload_date.append(upload.text)

ve = driver1.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for view in ve:
    Views.append(view.text)
```

```python
print(len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
import pandas as pd

print("Length of Name list:", len(Name))
print("Length of Artist list:", len(Artist))
print("Length of Upload_date list:", len(Upload_date))
print("Length of Views list:", len(Views))

if len(Name) == len(Artist) == len(Upload_date) == len(Views):
    data = {
        'Name': Name,
        'Artist': Artist,
        'Upload_date': Upload_date,
        'Views': Views
    }

    df = pd.DataFrame(data)
    print(df)
else:
    print("Error: Lists are of different lengths. Cannot create DataFrame.")
```

```python
driver1.close()
```

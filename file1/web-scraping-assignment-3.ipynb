```python
import selenium
import pandas as pd
import time
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
drivers1=webdriver.Chrome()
```

```python
url="https://www.amazon.in"
drivers1.get(url)
```

```python
user_input=input("Enter the Product you want to search for :")
```

```python
search=drivers1.find_element(By.ID,"twotabsearchtextbox")
search.send_keys(user_input)

search2=drivers1.find_element(By.XPATH,'//div[@class="nav-search-submit nav-sprite"]/span/input')
search2.click()
```

```python
#Question-2
```

```python
product_url=[]
start =0
end =3
for page in range(start,end):
    url1=drivers1.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url1:
        product_url.append(i.get_attribute("href"))
    nxt_but=drivers1.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_but.click()
    time.sleep(2)
```

```python
len(product_url)
```

```python
brand_name=[]
product_name=[]
price=[]
exp_delivery=[]
availability=[]



for urls in product_url:
    drivers1.get(urls)
    time.sleep(2)
    
    
    try:
        brand=drivers1.find_element(By.XPATH,'//a[@id="bylineInfo"]')
        brand_name.append(brand.text)
    except NoSuchElementException:
            brand_name.append('-')
            
            
    try:
        product=drivers1.find_element(By.XPATH,'//span[@id="productTitle"]')
        product_name.append(product.text)
    except NoSuchElementException:
            product_name.append('-')
            
            
            
    try:
        prc=drivers1.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(prc.text)
    except NoSuchElementException:
            price.append('-')
            
            
            
    try:
        avai=drivers1.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        availability.append(avai.text)
    except NoSuchElementException:
            availability.append('-')
            
            
            
    try:
        delivery=drivers1.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        exp_delivery.append(delivery.text)
    except NoSuchElementException:
            exp_delivery.append('-')
```

```python
print(len(brand_name),len(product_name),len(price),len(exp_delivery),len(availability))
```

```python
guitar=pd.DataFrame({})
guitar['Brand Name']=brand_name
guitar['Product Name']=product_name
guitar['Price']=price
guitar['Expected Delivery']=exp_delivery
guitar['Availability']=availability

guitar
```

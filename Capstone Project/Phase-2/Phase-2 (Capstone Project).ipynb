```python
crime_by_state=Analysis2.groupby('STATE').sum()

crime_types = ['MURDER', 'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER',
               'RAPE', 'CUSTODIAL RAPE', 'OTHER RAPE', 'KIDNAPPING & ABDUCTION',
               'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS',
               'KIDNAPPING AND ABDUCTION OF OTHERS', 'DACOITY',
               'PREPARATION AND ASSEMBLY FOR DACOITY', 'ROBBERY', 'BURGLARY', 'THEFT',
               'AUTO THEFT', 'OTHER THEFT', 'RIOTS', 'CRIMINAL BREACH OF TRUST',
               'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT',
               'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY',
               'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES',
               'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES',
               'CAUSING DEATH BY NEGLIGENCE', 'OTHER IPC CRIMES']

```

```python
crime_by_state = Analysis2.groupby('STATE').agg({'Literacy Rate':'mean', 'MURDER':'sum', 'ATTEMPT TO MURDER':'sum',
                                                'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER':'sum', 'RAPE':'sum', 
                                                'CUSTODIAL RAPE':'sum', 'OTHER RAPE':'sum', 'KIDNAPPING & ABDUCTION':'sum',
                                                'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS':'sum', 
                                                'KIDNAPPING AND ABDUCTION OF OTHERS':'sum', 'DACOITY':'sum',
                                                'PREPARATION AND ASSEMBLY FOR DACOITY':'sum', 'ROBBERY':'sum', 'BURGLARY':'sum', 
                                                'THEFT':'sum', 'AUTO THEFT':'sum', 'OTHER THEFT':'sum', 'RIOTS':'sum',
                                                'CRIMINAL BREACH OF TRUST':'sum', 'CHEATING':'sum', 'COUNTERFIETING':'sum', 
                                                'ARSON':'sum', 'HURT/GREVIOUS HURT':'sum', 'DOWRY DEATHS':'sum', 
                                                'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY':'sum',
                                                'INSULT TO MODESTY OF WOMEN':'sum', 'CRUELTY BY HUSBAND OR HIS RELATIVES':'sum',
                                                'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES':'sum', 
                                                'CAUSING DEATH BY NEGLIGENCE':'sum', 'OTHER IPC CRIMES':'sum'})

plt.figure(figsize=(20, 10))

for crime_type in crime_by_state.columns[1:]:
    plt.bar(crime_by_state.index, crime_by_state[crime_type], label=crime_type)

ax2 = plt.gca().twinx()
ax2.plot(crime_by_state.index, crime_by_state['Literacy Rate'], color='k', linestyle='-', marker='o', label='Literacy Rate')
ax2.set_ylabel('Literacy Rate (%)')

plt.xlabel('State')
plt.ylabel('Frequency of Crimes')
plt.title('Analysis of  Types of Crimes and Literacy Rate Across States')
plt.xticks(rotation=90)
plt.legend(loc='upper left', title='Crime Type')
ax2.legend(loc='upper right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter


total_crime_by_year = Analysis3.groupby('YEAR')['TOTAL CRIMES'].sum()

def lakhs_formatter(x, pos):
    return f'{x/100000:.0f}L'

plt.figure(figsize=(10, 6))
total_crime_by_year.plot(kind='line', marker='o', color='b')
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('Year')
plt.ylabel('Total Crime Count (in lakhs)')

plt.gca().yaxis.set_major_formatter(FuncFormatter(lakhs_formatter))

plt.grid(True)
plt.show()
```

# 2.4	Analysis of area vs overall crime.

```python
Analysis4=df1[['STATE','YEAR','TOTAL CRIMES','Area\n (In sq. km)']]
Analysis4
```

```python
x=Analysis4['Area\n (In sq. km)']
y=Analysis4['TOTAL CRIMES']

plt.figure(figsize=(10,6))
plt.scatter(x,y,color='g',alpha=0.5)
plt.title('Total Area Vs Total Crimes')
plt.xlabel('Total Area in Sqkm')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

```python
x = Analysis4['Area\n (In sq. km)']
y = Analysis4['TOTAL CRIMES']
crime_counts = Analysis4['TOTAL CRIMES'] / 1000  # Divide by 1000 to scale down for bubble size
state_names = Analysis4['STATE']

# Create a bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(x, y, s=crime_counts, alpha=0.5)

# Annotate each point with state names
for i, txt in enumerate(state_names):
    plt.annotate(txt, (x[i], y[i]), fontsize=8, ha='center', va='center')

plt.title('Area vs Overall Crime')
plt.xlabel('Area (sq. km)')
plt.ylabel('Total Crimes (in thousands)')
plt.grid(True)
plt.show()
```

```python
sort = Analysis4.groupby("STATE").sum().reset_index()[['STATE', 'TOTAL CRIMES']].sort_values(by='TOTAL CRIMES', ascending=False)

def lakhs_formatter(x, pos):
    return f'{x/100000:.0f}L'

plt.figure(figsize=(15, 10))
ax=sns.barplot(x=sort['STATE'], y=sort['TOTAL CRIMES'])
plt.xticks(rotation=90)
plt.title('Total Crimes across all states')
plt.xlabel('State')
plt.ylabel('Total Crimes (in lakhs)')

plt.gca().yaxis.set_major_formatter(FuncFormatter(lakhs_formatter))

for p in ax.patches:
    ax.annotate(f'{p.get_height()/100000:.1f}L', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=8, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.grid(True)
plt.show()
```

# 2.5 Analysis of Population vs overall Crime.

```python
Analysis5=df1[['STATE','YEAR','TOTAL CRIMES','POPULATION']]
Analysis5
```

```python
x=Analysis5['POPULATION']/100000
y=Analysis5['TOTAL CRIMES']

plt.figure(figsize=(10,8))
plt.scatter(x,y,color='y',alpha=0.5)
plt.title('Total Population Vs Total Crimes ')
plt.xlabel('Total Population(in Lakhs)')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

```python
x = Analysis5['POPULATION'] / 100000  
y = Analysis5['TOTAL CRIMES']

plt.figure(figsize=(10, 8))
plt.plot(x, y, color='b', marker='o', linestyle='-')
plt.title('Total Population Vs Total Crimes')
plt.xlabel('Total Population (in lakhs)')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

```python
years = Analysis5['YEAR']
population = Analysis5['POPULATION'] / 100000000  # Convert population to 10 crores
crimes = Analysis5['TOTAL CRIMES']

# Define the width of the bars
bar_width = 0.35

# Set the x locations for the bars
index = np.arange(len(years))

plt.figure(figsize=(12, 8))

# Plot total population
plt.bar(index, population, bar_width, label='Population (in 10 crores)', color='blue')

# Plot total crimes stacked on top of total population
plt.bar(index, crimes, bar_width, bottom=population, label='Total Crimes', color='red')

plt.title('Year-wise Total Population and Total Crimes')
plt.xlabel('Year')
plt.ylabel('Count (in 10 crores)')
plt.xticks(index, years)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
